Import dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

Data collection and preproceesing

```python
calories = pd.read_csv('calories.csv')
```

```python
calories.head()
```

```python
exercise = pd.read_csv('exercise.csv')
```

```python
exercise.head()
```

Combining the two dataframe

```python
calories_data = pd.concat([exercise, calories['Calories']], axis = 1)
```

```python
calories_data.head()
```

```python
calories_data.shape
```

```python
calories_data.info()
```

```python
calories_data.isnull().sum()
```

Converting text into numerical

```python
calories_data.replace({"Gender":{'male':0,'female':1}}, inplace=True)
```

Finding the positive and Negative correlation

```python
 correlation = calories_data.corr()
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(correlation, cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size':8}, cmap='Blues')
```

Separating features and target

```python
X = calories_data.drop(columns=['User_ID','Calories'], axis=1)
Y = calories_data['Calories']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Traning

```python
model = XGBRegressor()
```

```python
model.fit(X_train, Y_train)
```

Model evaluation

```python
train_data_prediction = model.predict(X_train)
```

```python
mae = metrics.mean_absolute_error(Y_train, train_data_prediction)
print("Mean absolute errror : ", mae)
```

```python
test_data_prediction = model.predict(X_test)
```

```python
mae1 = metrics.mean_absolute_error(Y_test, test_data_prediction)
print("Mean absolute error : ", mae1)
```
