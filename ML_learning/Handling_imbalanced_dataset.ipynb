This is highly imbalanced dataset
0->legit
1->fraud

```python
#seperating the data
legit = credit_data[credit_data.Class == 0]
fraud = credit_data[credit_data.Class == 1]
```

```python
print(legit.shape)
```

```python
print(fraud.shape)
```

Under sampling

BUILDING A SAMPLE DATASET

```python
legit_sample = legit.sample(n = 38)
```

```python
print(legit_sample.shape)
```

Concatedanate the dataframe of fraud and newly generated legit_sample


```python
new_dataset = pd.concat([legit_sample, fraud], axis = 0)
```
