creating matrix with random value

```python
random_matrix = np.random.rand(3, 3)
print(random_matrix)
```

```python
#matrix with random integer
random_int_matrix = np.random.randint(2, 10, size = (3, 3))
print(random_int_matrix)
```

```python
#matrix with all value 1
matrix_3 = np.ones((3,3))
print(matrix_3)
```

```python
#integer matrix with all value 1
matrix_3 = np.ones((3,3), dtype=int)
print(matrix_3)
```

Null matrix

```python
#matrix with all value 0
matrix_4 = np.zeros((3,3))
print(matrix_4)
```

```python
#integer matrix with all value 0
matrix_4 = np.zeros((3,3), dtype = int)
print(matrix_4)
```

Identity matrix

```python
identity = np.eye(3,3)
print(identity)
```

Transpose of a matrix

```python
rand_matrix = np.random.randint(2, 10, size = (3, 3))
print(rand_matrix)
transpose = np.transpose(rand_matrix)
print(transpose)
```

Matrix addition

```python
a = np.array([[1, 3],[4, 7]])
b = np.array([[3, 5],[2, 4]])

c = a + b
print(c)
```

Subtraction of matrix

```python
sub = a - b
print(sub)
```

Matrix multiplication

```python
x = 5
y = np.random.randint(10, size = (3, 3))
print(y)
```

```python
prod = np.multiply(x, y)
print(prod)
```

```python
a = np.random.randint(10, size = (3, 3))
b = np.random.randint(10, size = (3, 3))
print(a)
print(b)
```

```python
prod = np.multiply(a, b)
print(prod)
```

```python
a = np.random.randint(10, size = (3, 3))
b = np.random.randint(10, size = (3, 4))
print(a)
print(b)
```

```python
prod = np.dot(a, b)
print(prod)
```

element wise matrix multiplicaton

```python
a = np.random.randint(10, size = (3, 3))
b = np.random.randint(10, size = (3, 4))
print(a)
print(b)
```
